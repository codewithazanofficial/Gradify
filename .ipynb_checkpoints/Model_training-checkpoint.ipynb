```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_lifestyle_dataset.csv')
```

```python
df.head(5)
```

```python
df.drop('Stress_Level', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
1.423888/7.475800 	* 100  # coefficient of Variation
```

```python
corr = df.corr()
corr.GPA.sort_values(ascending = False)
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop(['Student_ID', 'GPA'], axis = 1)
y = df.GPA
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 5, shuffle = True, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(X_test)
```

```python
x = np.array([6.9,3.8,8.7,2.8,1.8]).reshape(1,-1)
x.shape
```

```python
model.predict(x)
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
 mae / X.Study_Hours_Per_Day.mean() * 100
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score = cross_val_score(LinearRegression(), X_train, y_train, scoring = 'r2', cv = 10, n_jobs = -1).mean()
```

```python
cv_score
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
results = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred,
    'Residuals': y_test - y_pred
})

# Pair plot
sns.pairplot(results[['Actual', 'Predicted']])
plt.show()
```
